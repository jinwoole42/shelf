### 함수의 사칙연산
함수의 더하기, 빼기, 곱하기, 나누기를 해보자  

```python
import numpy as np
import matplotlib.pyplot as plt
```

1. f(x) = x + 1, g(x) = 3x - 10 일때, 함수의 사칙연산을 해보고, 코드를 통해 맞는지 확인해보자

```python
def f(x):
    return -3*x;
def g(x):
    return 3*x - 10;

x = np.linspace(-20, 20, 1000)
plt.plot(x, f(x)) #blue
plt.plot(x, g(x)) #orange
#plt.plot(x, f(x) + g(x)) #green -> 바꿔가면서, 직접 계산한 식이 맞나 확인해보자
plt.xlim([-20, 20])      # X축의 범위: [xmin, xmax]
plt.ylim([-20, 20])     # Y축의 범위: [ymin, ymax]
plt.axvline(x=0, color = 'r')  # x=0축
plt.axhline(y=0, color = 'r')   # y=0축
plt.show()
```

2. 함수의 평행이동을 해보자. $y = 2x^2$ - 3$ 일때, x축 방향으로 5 만큼, y축 방향으로 1만큼 평행이동한 그래프를 그려라

```python
def f(x):
    return 2*x*x - 3;
def g(x):
    return 2*(x-5)*(x-5) - 2;
plt.xlim([-7, 7])      # X축의 범위: [xmin, xmax]
plt.ylim([-7, 7])     # Y축의 범위: [ymin, ymax]
plt.axvline(x=0, color = 'r')  # x=0축
plt.axhline(y=0, color = 'r')   # y=0축
plt.axvline(x=5, color = 'g')  # x=0축
plt.axhline(y=-2, color = 'g')   # y=0축
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.show()
```

3. 함수의 대칭이동을 해보자. 새로 뭐 하기 귀찮으니, 위의 g(x)를 각각 x축, y축, 원점에 대해 각각 대칭이동해보자  

```python
plt.plot(x, -g(x)) ##x축에 대한 대칭이동
plt.xlim([-7, 7])      # X축의 범위: [xmin, xmax]
plt.ylim([-7, 7])     # Y축의 범위: [ymin, ymax]
plt.axvline(x=0, color = 'r')  # x=0축
plt.axhline(y=0, color = 'r')   # y=0축
plt.show()
```

```python
plt.plot(x, g(-x)) ##x축에 대한 대칭이동
plt.xlim([-7, 7])      # X축의 범위: [xmin, xmax]
plt.ylim([-7, 7])     # Y축의 범위: [ymin, ymax]
plt.axvline(x=0, color = 'r')  # x=0축
plt.axhline(y=0, color = 'r')   # y=0축
plt.show()
```

```python
plt.plot(x, -g(-x))
plt.xlim([-7, 7])      # X축의 범위: [xmin, xmax]
plt.ylim([-7, 7])     # Y축의 범위: [ymin, ymax]
plt.axvline(x=0, color = 'r')  # x=0축
plt.axhline(y=0, color = 'r')   # y=0축
plt.show()
```

4. $y = \sqrt{x} + 1$ 함수를 그리고, 그걸로 대칭이동과 평행이동을 해보자  

```python
x = np.linspace(0, 20, 200)

def f(x):
    return np.sqrt(x) + 1;
plt.plot(x, f(x))
plt.xlim([-7, 7])      # X축의 범위: [xmin, xmax]
plt.ylim([-7, 7])     # Y축의 범위: [ymin, ymax]
plt.axvline(x=0, color = 'r')  # x=0축
plt.axhline(y=0, color = 'r')   # y=0축
plt.show()
```
